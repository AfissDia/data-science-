```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://www.ucad.sn"
req = requests.get(url)

if(req.status_code==200):
    print(req.content)
else:
    print("Not Ok")
```

```python
url = "https://www.ucad.sn"
req = requests.get(url)

if(req.status_code==200):
    soup = BeautifulSoup(req.content,'html.parser')
    all_link = soup.find_all('a')
    for link in all_link:
        print(link.get("href"))
else:
    print("Not Ok")
```
