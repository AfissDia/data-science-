```python
df_PL_2021_2021_matchs = pd.read_csv('DATAS/PL_2020_2021_matchs.csv')
df_PL_2021_2021_matchs
```

```python
df_PL_2021_2021_matchs['Date_match']=pd.to_datetime(df_PL_2021_2021_matchs['Date'])
```

```python
df_PL_2021_2021_matchs['Date_match']=df_PL_2021_2021_matchs['Date_match'].dt.month
```

```python
df_man_city = df_PL_2021_2021_matchs[(df_PL_2021_2021_matchs['HomeTeam']=='Man City')]
df_man_city.head()
```

```python
df_man_city_home = df_man_city[df_man_city['FTR']=='H']
df_man_city_home.head()
```

```python
df_city_home = df_man_city_home.groupby(['HomeTeam','FTR','Date_match'])['FTR'].count().rename('nbr_match').reset_index()
df_city_home
```

```python
df_man_city_Away = df_PL_2021_2021_matchs[(df_PL_2021_2021_matchs['AwayTeam']=='Man City')]
```

```python
df_city_Away = df_man_city[df_man_city['FTR']=='A']
df_city_Away
```

```python
df_city_Away = df_city_Away.groupby(['AwayTeam','FTR','Date_match'])['FTR'].count().rename('nbr_match').reset_index()
df_city_Away
```

```python
match_gagner = df_city_home.merge(df_city_Away,on=('nbr_match'))
match_gagner
```
