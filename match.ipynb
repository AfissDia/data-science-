### afficher l'equipe qui a remporter le championat

```python
domicile=pl_2021_2021_matchs.groupby(['HomeTeam','FTR'])['FTR'].count().rename('resultat').reset_index()
domicile.head()
```

```python
gagner_domicile= domicile[domicile['FTR']=='H']
gagner_domicile.head()
```

```python
exterieur=pl_2021_2021_matchs.groupby(['AwayTeam','FTR'])['FTR'].count().rename('resultat').reset_index()
exterieur.head()
```

```python
gagner_exterieur= exterieur[exterieur['FTR']=='A']
gagner_exterieur.head()
```

```python
match_null_domicile=domicile[domicile['FTR']=='D']
match_null_domicile.head()
```

```python
match_null_exterieur=exterieur[exterieur['FTR']=='D']
match_null_exterieur
```

```python
match_null_domicile.rename(columns={'HomeTeam' : 'team'},inplace=True)
match_null_exterieur.rename(columns={'AwayTeam' : 'team'},inplace=True)
```

```python
match_null_domicile['points']=match_null_domicile['resultat']
match_null_exterieur['points']=match_null_exterieur['resultat']
```

```python
match_null = match_null_domicile.merge(match_null_exterieur, on =('team'))
match_null
```

```python
match_null['nbr_match_null']=match_null['resultat_x'] + match_null['resultat_y']
match_null
```
