# Exercice championat Angleterre 2020 2021

```python
df_PL_2021_2021_matchs = pd.read_csv('DATAS/PL_2020_2021_matchs.csv')
df_PL_2021_2021_matchs
```

## nbr de points gagner a exterieur

```python
# nombre de points a exterieur
match_exterieur = df_PL_2021_2021_matchs.groupby(['AwayTeam','FTR'])['FTR'].count().rename('nbr_match').reset_index()
match_exterieur.head()
```

```python
match_gagner_exterieur = match_exterieur[match_exterieur['FTR']=='A']
match_gagner_exterieur.head()
```

```python
match_gagner_exterieur['nbr_points']=match_gagner_exterieur['nbr_match']*3
match_gagner_exterieur
```

## nbr de points gagner a domicile

```python
# nombre de points a exterieur
match_domicile = df_PL_2021_2021_matchs.groupby(['HomeTeam','FTR'])['FTR'].count().rename('nbr_match').reset_index()
match_domicile.head()
```

```python
match_gagner_domicile = match_domicile[match_domicile['FTR']=='H']
match_gagner_domicile.head()
```

```python
match_gagner_domicile['nbr_points']=match_gagner_domicile['nbr_match']*3
match_gagner_domicile
```

```python
match_gagner_exterieur.rename(columns={'AwayTeam':'team'}, inplace=True)
match_gagner_domicile.rename(columns={'HomeTeam':'team'}, inplace=True)
```

# dataframe nbr points a domicile et exterieur

```python
match_gagner = match_gagner_domicile.merge(match_gagner_exterieur, on=('team'),suffixes=('_domicile','_exterieur'))
match_gagner
```

```python
match_gagner['nbr_points_gagner'] = match_gagner['nbr_points_domicile'] + match_gagner['nbr_points_exterieur']
match_gagner
```

# les matchs null exterieur et interieur

```python
# match null dimicile
match_null_domicile = match_domicile[match_domicile['FTR']=='D']
match_null_domicile.head()
```

```python
# match null exterieure
match_null_exterieure = match_exterieur[match_exterieur['FTR']=='D']
match_null_exterieure.head()
```

```python
match_null_domicile.rename(columns={'HomeTeam':'team'},inplace=True)
match_null_exterieure.rename(columns={'AwayTeam':'team'},inplace=True)
```

```python
#dataframe match null
df_match_null=match_null_domicile.merge(match_null_exterieure,on='team',suffixes=('_dom','_ext'))
df_match_null
```

```python
df_match_null['nbr_points_null']=df_match_null['nbr_match_dom']+df_match_null['nbr_match_ext']
df_match_null
```

```python
# nbr point total
resultat_des_matchs = match_gagner.merge(df_match_null, on='team')
resultat_des_matchs
```

```python
# nombre de but a exterieur
resultat_des_matchs['nbr_points_final'] = resultat_des_matchs['nbr_points_gagner']+resultat_des_matchs['nbr_points_null']
resultat_des_matchs
```

```python
# Classement par ordre le plus de points
resultat_des_matchs.sort_values(by='nbr_points_final', ascending=False)
```
